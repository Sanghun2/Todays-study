```python
my_groups = data.groupby('Region')
my_names = []
for ( name, dfb ) in my_groups:
    my_names += [name]
```

```python
my_names
```

```python
my_groups = data.groupby('Region')
my_names = []
my_list = []
for ( name, dfb ) in my_groups:
    my_names += [name]
    my_list += [dfb.Units.values]
```

```python
plt.boxplot(my_list, False, labels=my_names) # data자리에는 리스트나 넘피배열이 들어가야 한다.
plt.show()
```

```python
my_list
```

```python
my_groups = data.groupby('Region')
my_names = []
my_list = []
for ( name, dfb ) in my_groups:
    my_names += [name.upper()] # 대문자로 변경
    my_list += [dfb.Units.values]
```
