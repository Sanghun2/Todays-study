# 파이썬 반복문과 함수
파이썬의 반복문은 여러가지가 있다.<br><br>
**FOR**<br><br>
**WHILE**<br><br>
<br><br>
for문과 while문을 컨트롤하는 방법과 차이를 적어보려고 한다.
여기서 말할 것은 RETUEN과 BREAK다. 전에는 분별없이 사용했었는데 단순히 빠져나가고 싶으면 break나 return둘 중 하나만 쓰면 되는줄 알았다. 그래서 함수에서 break를 쓰고 반복문에서 return을 쓰기도 했었다. 하지만 그게 아니었고 큰 차이가 있었다.<br><br>
<br><br>
**RETURN**의 경우 함수안에서만 사용한다.
**BREAK**는 반복문을 멈출 때 사용한다.
<br><br>
이것이 어떤 차이냐 하면 함수안에 있는 For문을 예로 들어보면 For문 안에 break를 쓰면 For문이 종료되고 함수 내부의 다음 명령을 수행한다. 하지만 For문안에 RETURN이 수행되면 그대로 함수값은 Return값이 되고 그대로 함수가 종료된다. 차이가 분명하다. Break는 함수값을 반환하지 않고 반복문만 종료되지만 return은 반환값을 반환하고 함수가 종료된다.
이 차이를 분명히 이해하고 사용해야 한다.<br><br>

# 파일 읽고 쓰기
<br><br>
**file_data = open('file.txt')**
**print(file_data.readline(), end="")**
**file_data.close()**<br><br>
<br><br>
**with open('file.txt') as file_data:**
    **print(file_data.readline(), end="")**<br><br>
<br><br>
원문은 위의 문장이지만 with문을 사용하면 아래처럼 좀 더 단순하게 사용할 수 있다.
원칙적으로 파일을 열면 close()를 써줘야 하지만 with문을 사용할 때는 with문을 나올 때 자동으로 close를 불러주기 떄문에 따로 쓰지 않아도 된다.
<br><br>
### 쓰기(Write)
<br><br>
**with open('file.txt', 'w') as file_data:**
    **file_data.write("First\n")**
    **file_data.write("Second")**<br><br>
<br><br>
<br><br>
쓰기모드는 'w'로 명시하며 이 때 파일.write("문장")을 하게되면 해당 파일에 쓸때마다 계속 덮어쓰이게 된다.
<br><br>
### 읽기(Read)
<br><br>
**with open('file.txt', 'r') as file_data:**
    **for line in file_data:**
        **print(line)**<br><br>
        <br><br>
읽기모드는 'r'로 쓰며 데이터를 읽어오는 것만 가능하다.
이미 적혀있는 데이터를 위 문장처럼 for문으로 읽어오면 파일의 리스트가 조건으로 들어가게 되고 for문을 돌며 차례로 읽어와서 print를 통해 보여진다.
<br><br>
<br><br>
### 추가(Append)
<br><br>
**with open('file.txt', 'a') as file_data:**
    **file_data.write("Third\n")**<br><br>
<br><br>
읽기모드는 'a'로 쓰며 기본적인 방법은 write와 같지만 차이점이 있다면 w와 달리 a는 기록할때마다 덮어쓰는 것이 아니라 계속 추가가 된다는 것이다.

클래스는 일종의 틀이라고 할 수 있다. 변수나 함수등을 클래스에 정의해 놓으면 객체를 만들었을때 그 변수와 함수를 그대로 사용할 수 있다.
같은 작업을 반복하거나 생성된 객체들을 관리해야 할 때 만약 클래스가 아니라 하나씩 작업했다면 다시 하나씩 수정해야 하지만 클래스로 만든 객체는 클래스 부분만 수정해주면 모든 객체가 바뀌기 때문에 유지,보수 면에서도 효율성이 엄청나다고 할 수 있다. 정리해 보자면 클래스의 큰 장점은 다음 2가지라고 할 수 있을 것 같다.<br><br>

이처럼 멤버 변수는 멤버 메소드 안에서 얼마든지 바꿀 수 있고, 그 영향이 해당 메소드안에서만 미치지만, 클래스 변수는 전체에 영향을 미친다.

### 예외처리
<br><br>
>##### 예외처리가 필요한 이유
>> 프로그램에 오류가 발생하거나 발생하지 않았을 때에 대한 처리를 해줘야 프로그램이 유연하게 돌아갈 수 있다.
<br><br>
>##### 예외처리문의 종류
>###### 모든에러처리
>> try:<br><br>
>><예외 발생가능 코드><br><br>
>>except:<br><br>
>><예외 발생시 수행되는 코드>
<br><br>

>###### 특정에러처리
>> try:<br><br>
>><예외 발생가능 코드><br><br>
>>except <발생 에러>:<br><br>
>><예외 발생시 수행되는 코드>
<br><br>
except문에서 "<발생에러> as 변수" 형태를 통해 발생한 에러를 변수형태로 저장하여 print등에 이용할 수 있다.
<br><br>

>###### 에러처리 및 에러가 발생하지 않는 경우
>> try:<br><br>
>><예외 발생가능 코드><br><br>
>>except:<br><br>
>><예외 발생시 수행되는 코드><br><br>
>>else:<br><br>
>><예외 발생시 수행되는 코드><br><br>
>>finally:<br><br>
>><예외 발생 여부와 상관없이 항상 수행되는 코드><br><br>
<br><br>
