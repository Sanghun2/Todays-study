## 필터링 (WHERE)
#### (AND, OR, NOT)
**SELECT name from country WHERE population < 1000 AND continent = 'Asia' ORDER BY population DESC;**<br><br>
  <br><br>
이 문장의 SELECT 순서는 population 조건을 만족하고 continent를 만족하는 것의 name을 출력하고 population을 기준으로 내림차순 정렬하는 것을 의미한다.  
위처럼 WHERE을 사용해 출력할 데이터의 조건을 설정할 수 있다.
  
AND를 쓰면 양 쪽 조건을 만족하는 데이터
OR은 둘 중에 하나라도 만족
NOT은 조건을 만족하지 않는 데이터를 출력한다.
  
#### 쓸 수 있는 비교연산자
**a = b**	a와 b는 같다
**a <> b**	a와 b는 같지 않다
**a > b**	a는 b보다 크다
**a >= b**	a가 b 이상이다
**a < b**	a는 b보다 작다
**a <= b**	a가 b 이하이다
  
같은지 같지 않은지에 대해서는 다음과 같은 연산자도 사용할 수 있다.
  
비교 연산자	설명
**a == b**	a와 b는 같다.
**a != b**	a와 b는 같지 않다.
  
#### BETWEEN ~ AND ~
BETWEEN 다음 숫자를 쓰고 AND 숫자를 쓰면 양 쪽 숫자 범위안의 데이터를 출력
  
#### LIKE
**SELECT name from country WHERE name LIKE '%land' **
    
%land를 하면 앞에 무슨 문자가 있든 뒤에 land로 끝나는 데이터를 가져올 수 있다.
%land 양쪽에 반드시 ' '를 붙여야 한다.
  
**'land%'**   -- land로 시작하는 데이터
**'%land%'**  --  중간에 land가 들어가 있는 데이터
**'_a%' **    -- 두번째 문자가 a인 데이터
  
%는 불특정한 문자열을 의미하고, _는 불특정 문자를 의미

## FOREIGN KEY
<br><br>
**FOREIGN KEY 컬럼1 REFERENCES 테이블(컬럼2)**<br><br>
<br><br>
테이블의 컬럼2를 현재 테이블의 컬럼1에 foreign key로 연결한다.  
foreign key로 연결되는 컬럼2의 경우 중복되면 안되기 때문에 Primary key이며 컬럼1이 있는 테이블은 컬럼2에 없는 데이터가 추가될 수 없다. 일종의 제약역할이다.  
다시말해 컬럼1에 데이터를 넣으려면 넣으려는 값이 컬럼2에 있어야만 가능하다.  
또 데이터를 삭제할 때 foreign key로 이용되는 테이블은 먼저 삭제가 불가하지만 foreign key를 이용한 테이블은 먼저 삭제 할 수 있다.

4. 테이블에서 조건을 만족하는 컬럼의 레코드 중 최소와 최대를 계산

## GRUOP BY
<br><br>
**select 컬럼1, sum(컬럼2)**<br><br>

**from 테이블**<br><br>

**where 조건1**<br><br>

**group by 컬럼3**<br><br>

**having 조건2**<br><br>

**order by 컬럼4;**<br><br>
<br><br><br><br>
이 문장은 테이블에서 조건1을 만족하는 데이터 중 group by에 나온 컬럼을 기준으로 그룹을 만들고 조건2에 맞는 데이터를 선택해 컬럼4를 기준으로 출력한다는 의미.  
GROUP BY를 쓰면 그 컬럼을 기준으로 겹치는 것을 그룹으로 묶겠다는 의미이고, 합을 구하거나 평균 등을 구할 때 해당 그룹의 합과 평균을 각각 알 수 있다.  
HAVING은 조건문인데 WHERE이 SELECT할 데이터의 조건인 반면 HAVING은 조건에 포함되는 데이터만 Group by되게 한다.

**conn = sqlite3.connect('customer.db')**   <br><br>   - 데이터베이스에 접근 및 생성<br><br><br><br>
**cur = conn.cursor()**              <br><br>         - 커서 객체 생성<br><br><br><br>
**cur.execute(" sql 문장 ")**        <br><br>     - 커서객체로 데이터베이스에서 쿼리문 실행<br><br><br><br>
**cur.execute(" INSERT INTO 테이블 VALUES (?,?,?) ")**   <br><br>- ?를 사용해 사용자에게 입력값을 받는다<br><br>
**cur.executemany(" INSERT INTO 테이블 VALUES (?,?,?),**<br><br> - executemany를 사용하면 여러개의 값을 받는다.<br><br>
**(1001, 5, 10),**<br><br>
**(1002. 7, 11)**<br><br>
**);**<br><br>
**conn.commit()**                                            ----------- commit<br><br>
**conn.close()**                                             ----------- 데이터베이스 연결종료<br><br>
