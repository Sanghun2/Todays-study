```python
import pandas as pd

data = pd.read_csv(r'C:\Apps\temp/Traffic_Accident_2017.csv', encoding = 'euc-kr')
data.T #행과 열을 나눠서 보기
```

```python
from matplotlib import font_manager, rc #rc를 가져와야 rc모듈을 사용해 불러온다.결론적으로는 둘다 가져와야함

font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
```

```python
# data['요일'].value_counts()

import matplotlib.pyplot as plt

t = data['요일'].value_counts()
y = t[['월','화','수','목','금','토','일']]
x = ['월','화','수','목','금','토','일']
plt.bar(x, y)
```

```python
plt.ylim(500, 620)

plt.xlabel("요일")
plt.ylabel("사고건수")
plt.title("요일별 교통사고 사망 건수")

plt.bar(x, y)
plt.show()
```

```python
#차대차 사고 지역건수
data.T #g행과 열을 교차해서 볼 수 있음

#사고유형_대분류는 차대차인 지역의 정보
#pandas로 쿼리문을 만든다고 생각하면 됨.


t = data[['사고유형_대분류','발생지시도']][data['사고유형_대분류'] == '차대차']

x = t['발생지시도']
y = t['사고유형_대분류'].value_counts()
plt.ylim(1600, 1700)
plt.bar(x, y)
plt.show()
```

```python
# data.T
# t = data['발생년월일시'].value_counts()
# x = data['발생년월일시']
# y = t[]

data['발생년월일시']%100
t = [x % 100 for x in data['발생년월일시']]
time_df = pd.DataFrame(t, columns = ['사고시간'])
time_df
```

```python
# 카테고리
ages = [14,15,16, 23, 37, 31, 61, 20, 41, 32, 100]
bins = [0, 15, 25, 35, 60, 99]
labels = ["미성년자", "청년", "중년", "장년", "노년"]
cats = pd.cut( ages, bins, labels=labels)
cats

#NaN은 
```

```python
bins = [0, 2, 5, 8, 11, 14, 17, 20, 24]
labels = ["0-2", "3-5", "6-8", "9-11", "12-14", "15-17", "18-20", "21-23"]
time_cats = pd.cut(time_df['사고시간'], bins, labels=labels)
rs = time_cats.value_counts()

plt.pie(rs, labels = labels, autopct = '%1.1f%%')
plt.show
```
