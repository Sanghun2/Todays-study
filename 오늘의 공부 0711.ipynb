## git에 관하여

### 저장소(repository)
저장소는 데이터를 저장할 수 있는 공간을 말한다.
저장소의 종류는 로컬저장소와 원격 저장소가 있다.

### 로컬을 만드는 방법
로컬 저장소를 만들때는 직접만드는 방법과 이미 만들어져 있는 저장소를 복사 해오는 방법이 있다.

### Commit
commit을 하면 이전 커밋부터 지금까지의 변경이력이 기록된 커밋이 만들어진다.
순서대로 저장되고 과거이력과 내용을 알 수 있다.
각 commit에는 영문+숫자로 된 40자리의 고유이름이 붙는다.
commit을 할 때 메세지를 입력하지 않으면 commit되지 않는다.

### 작업트리?
git에서는 폴더를 작업트리라고 한다.

### Stage?
git에서 commit할 때 바로 이동되지 않고 중간에 인덱스라는 가상공간을 거치게 되는데 각 파일들은 인덱스를 가지게 된다.
인덱스에서 올릴 파일을 선택하는 작업을 스테이징(staging)이라고 한다.

## Git 주요 용어 및 구성

#### Git?
버전 관리를 위한 프로그램. VCS (Version Control System). 버전을 효율적으로 관리할 수 있도록 도와준다.
reote repository에 올려서 협업을 하거나 독립적으로 버전을 관리할 수 있다.

#### 버전?
각 날마다 공부한 내용이 담긴 파일이 있을때 각 날마다 공부한 내용을 버전이라고 한다.

### <영역>

#### Working directory
현재 작업하고 있는 공간
git이 관리하고 있긴하지만 아직 추적(tracking)하고 있지는 않다.

#### Index
임시 저장공간, 추적하고 있으며 버전으로 등록되기 전 상태.
stage, stage area

#### Repository
저장소
로컬 저장소와 원격 저장소가 있다.

### < Flow>

#### git init
새로운 local repository (.git 폴더를 생성)를 생성
git폴더가 있어야 추적이 가능하며, git 관련 작업을 할 수 있다.

#### git add
working area의 변경된 내용을 staging area로 추가시킨다.

#### git commit
staging area에 저장된 내용을 local repository로 확정시킨다.

#### git push
local repository에 있는 내용을 remote repository로 업로드.

#### git pull
local repository에 있는 내용을 remote repository에서 다운로드.

#### git clone
.git을 갖고있는 remote repository의 파일들을 local repository에 복사.
깃헙에서 zip파일로 받으면 .git 폴더가 없다는 것이 명령어와의 차이점.

### <협업-병합>
#### git branch
독립된 working directory.
참여자마다 독립된 작업공간을 가진다.
테스트 및 백업의 용도로 사용가능

#### head.?
포인터.
현재 작업하고 있는 branch를 가르킴

#### merge
여러개의 branch에서 작업한 내용을 하나로 합치는 것
현재 branch를 기준으로 병합.
두 branch가 같은 파일의 같은 곳을 작업했다면 충돌 발생 (merge conflict).
    관계자들이 상의하여 수동으로 해결.
    충돌 이슈가 발생하지 않으려면 작업 내용이 겹치지 않도록 분리하는 것이 좋다.

 ## Github에 소스코드 올리기
 
 ### 처음 코드 올릴 때
 
 #### 레포지토리 생성
 github에 repository를 생성한다.
 
 #### 업로드할 폴더로 이동
 cd 폴더경로
 
    업로드할 파일이 있는 폴더로 이동한다.
 
 #### init
 git init
 
    git이 폴더를 추적할 수 있도록 .git폴더를 생성한다.(local repository 생성)
 
 #### 상태확인
 git status
 
     git이 버전관리 대상 파일의 상태를 확인.
     명령어가 동작하지 않을 때 에러를 확인하거나 작업한 파일외의 파일이 수정되진 않았는지 확인.
 
 #### add
 git add .

    버전 관리할 모든 파일을 추가
    
 git add 파일명

     특정 파일을 추가
    
 #### commit
 git commit -m" "
 
     -m은 한줄로 메세지를 작성하기 위한 명령어이고 긴 문장을 작성하려면 git commit만 쓰면 된다.
 
#### remote 등록
git remote add origin 원격레포지토리주소

    origin은 remote repository의 별칭을 의미.
    
#### push
git push origin 브렌치명

    master는 branch의 이름. remote repository를 생성하면 기본적으로 master branch가 생김.
    브렌치는 독립적인 작업공간을 의미.
    
    
#### config
git을 설치하고 설정을 하지 않았다면 github 이메일 주소 및 비밀번호를 입력하라고 한다.

### 다시 코드 올릴 때
git add .
git commit -m"메시지 내용"
git push origin 브렌치이름

# git을 배우며...

2일전 git이라는 것을 처음 배웠다. 강사님에게 피드백을 드릴때 처음하는 것이 많아 양이 많은 것은 괜찮은데 무엇부터 해야할지 모르겠다고 했었는데 다음날 우리반만 git이나 vs code같은 것들을 필요하다면서 알려주셨다. 
어제는 시험을 보고 뻗어서 이제서야 제대로 한 번 보는데 깊이가 깊고 양이 방대한 것 같다.
오늘은 강사님께서 알려주셨던 것을 따라가보면서 관련된 개념을 정리하며 나가보고자 한다. 파이썬 클래스도 아직 제대로 만들 줄 몰라서 시간을 두고 차근차근 해나가야 지치지 않고 할 수 있을 것 같다.

강사님이 알려주신 내용은 바로 깃을 이용해 vs code에서 commit하는것

현재 대충 기억나는 것은 내용을 작성하고 git init부터 시작해서 git push origin master를 하여 최종 업로드를 하는 것이다.

잠깐 공부를 했을 때 얻은 깃의 흐름은 처음 올리면 인덱스라고 하는 가상공간을 거쳐 선택된 파일들이 서버로 올라가는 것 같다.
